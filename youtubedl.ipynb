```python
def creat_folder(URL):  
  #
  with youtube_dl.YoutubeDL({"ignoreerrors": True, "quiet": True}) as ydl:
      playlist_dict = ydl.extract_info(URL, download=False)
      print('\n', playlist_dict['title'], '\n')
  #
  new_folder = playlist_dict['title'] + '_' + playlist_dict['uploader']
  new_folder = new_folder.translate(table_punctuation).translate(table_digit) 
  #
  path = f"//content//drive//MyDrive//YouTube//{new_folder}//"
  os.makedirs(path, exist_ok=True)
  #
  os.chdir(path)
  print(os.getcwd())

  with open(f"_url.txt", "w", encoding="utf-8") as f:
          f.write(URL)
```

```python
def download_video(URL, playlist_start=1):
    youtube_dl_options = {
        "format": "(bestvideo+bestaudio/best)", 
        "outtmpl":"%(autonumber)s_%(title)s.%(ext)s",  
        "autonumber_start": playlist_start,
        "playliststart": playlist_start,    
        "restrictfilenames": False,
        "nooverwrites": True,
        "writedescription": True,
        "writeinfojson": False,
        "writeannotations": True,
        "writethumbnail": False,
        'writesubtitles': False,
        'writeautomaticsub': False,
        'subtitlesformat': "srt/ass/best"
    }

    is_not_end = True
    while(is_not_end):
      try:
        with youtube_dl.YoutubeDL(youtube_dl_options) as ydl:
            ydl.cache.remove()
            info_dict = ydl.extract_info(URL, download=False)
            ydl.prepare_filename(info_dict)
            ydl.download([URL])
            is_not_end = False
            return True
      except Exception:
        print('*'*50)
        is_not_end = True
```

```python
code = 'a.en'

def float_to_srt_time_format(d: float) -> str:

        fraction, whole = math.modf(d/1000)
        time_fmt = time.strftime("%H:%M:%S,", time.gmtime(whole))
        ms = f"{fraction:.3f}".replace("0.", "")
        return time_fmt + ms

def xml_caption_to_srt(xml_captions: str) -> str:

        segments = []
        root = ElementTree.fromstring(xml_captions)
        count_line = 0
        for i, child in enumerate(list(root.findall('body/p'))):
        
            text = ''.join(child.itertext()).strip()
            if not text:
                continue
            count_line += 1
            caption = unescape(text.replace("\n", " ").replace("  ", " "),)
            try:
                duration = float(child.attrib["d"])
            except KeyError:
                duration = 0.0
            start = float(child.attrib["t"])
            end = start + duration
            try:
                end2 = float(root.findall('body/p')[i+2].attrib['t'])
            except:
                end2 = float(root.findall('body/p')[i].attrib['t']) + duration
            sequence_number = i + 1  # convert from 0-indexed to 1.
            line = "{seq}\n{start} --> {end}\n{text}\n".format(
                seq=count_line,
                start=float_to_srt_time_format(start),
                end=float_to_srt_time_format(end2),
                text=caption,
            )
            segments.append(line)

        return "\n".join(segments).strip()
```

```python
def download_srt(URL):
  if 'playlist?list=' in URL.lower():
    playlist = Playlist(URL).video_urls
  else:
    playlist = []
    playlist.append(URL)
  for i, video_url in  enumerate(playlist):
    # create an object of YouTube() and pass the  URL of YouTube Videos
    video_src = YouTube(video_url)
    # to get particular langauge caption you need to pass the language code e.g, captions['a.en']
    if code in video_src.captions:
      en_caption_data = video_src.captions[code]

      #encode in xml format
      xml_caption = en_caption_data.xml_captions 

      xml_caption_to_srt_en = xml_caption_to_srt(xml_caption) 
      
      #saving the the captions to a flat file
      title = video_src.title.translate(table_punctuation).translate(table_digit) 

      #saving the subtitle in the file   
      j = str(i+1).zfill(5)
      print(j,' ', title)
      with open(f"{j}_{title}.srt", "w", encoding="utf-8") as f:
          f.write(xml_caption_to_srt_en)
```

```python
for URL in URLS:
  if isinstance(URL, dict) and len(URL)==1:
    url = list(URL.keys())[0]
    playlist_start =  list(URL.values())[0]
  elif isinstance(URL, str):
    url = URL
    playlist_start =  1
  else:
    print('There are some errors in URLS')
    break    

  creat_folder(url)
  print('-'*50)

  download_video(url, playlist_start)
  print('-'*50)
  
  download_srt(url)
  print('-'*50)
```
